```python
import pandas as pd
import sys
import os
```

```python
xls_files = os.listdir('../../temp_data')
```

```python
xls_files
```

```python
for files in xls_files:  
    xls = pd.ExcelFile('../../temp_data/'+files)
    sheets = xls.sheet_names
    if len(sheets) == 1:
        xls.parse(sheets[0]).to_csv(files[:-4] + '.csv', encoding = 'utf-8')
    else:
        final_df = xls.parse(sheets[0])
        headers = final_df.columns
        for i in range(1, len(sheets)):
            temp_df = xls.parse(sheets[i], header=None, names=headers)
            final_df = pd.concat([final_df, temp_df])
        final_df.to_csv(files[:-4] + '.csv', encoding = 'utf-8')
```
